## Melting/Unpivoting a DataFrame

#### Client has data for 30 individuals pizza consumption each year from 1990-2005. They would like to change the format of the DataFrame from wide to long.

#### Import Pandas as pd

```python
import pandas as pd
```

#### Import and assign DataFrame

```python
df = pd.read_csv('melt_fake_data.csv')
```

```python
df.head()
```

```python
#drop id column
df = df.drop(['id'], axis=1)
```

#### id_varse column(s) are used as identifier variables

```python
df.melt(id_vars=['first_name', 'last_name'])
#id_varse column headers will stay
```

#### Set column names

```python
df = df.melt(id_vars=['first_name', 'last_name'], var_name ='year', value_name='number_of_pizzas_eaten')
```

#### Result

```python
df
```

#### Export to excel

```python
df.to_excel('melted_fake_data.xlsx')
```
